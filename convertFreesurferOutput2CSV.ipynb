```python
freeOut = pd.DataFrame(columns=freeOut_columns)
for subject in os.listdir(freesurferOutputPath):
    print(subject)
    tempColumnDict = {}
    for root, dirs, files in os.walk(os.path.join(freesurferOutputPath,subject)):
        for name in files:
            if name == "lh.aparc.stats":
                path = os.path.join(root,name)
                with open(path) as f:
                    lines = f.readlines()
                    for v in freeOut_columns[1:]:
                        r2f = v.split('_')[1].lower()
                        region_name = v.split('_')[2].lower()
                        vol2thck = v.split('_')[3]
                        for line in lines:
                            lineList = line.split()
                            if region_name == lineList[0] and r2f == name[0]:
                                if vol2thck == 'Vol':
                                    tempColumnDict[v] = np.float64(lineList[3])
                                if vol2thck == 'Area':
                                    tempColumnDict[v] = np.float64(lineList[2])
                                if vol2thck == 'Thck':
                                    tempColumnDict[v] = np.float64(lineList[4])
            if name == "rh.aparc.stats":
                path = os.path.join(root,name)
                with open(path) as f:
                    lines = f.readlines()
                    for v in freeOut_columns[1:]:
                        r2f = v.split('_')[1].lower()
                        region_name = v.split('_')[2].lower()
                        vol2thck = v.split('_')[3]
                        for line in lines:
                            lineList = line.split()
                            if region_name == lineList[0] and r2f == name[0]:
                                if vol2thck == 'Vol':
                                    tempColumnDict[v] = np.float64(lineList[3])
                                if vol2thck == 'Area':
                                    tempColumnDict[v] = np.float64(lineList[2])
                                if vol2thck == 'Thck':
                                    tempColumnDict[v] = np.float64(lineList[4])
            if name == "aseg.stats":
                path = os.path.join(root,name)
                with open(path) as f:
    #                 for line in f:
    #                     print(line.split())
                    lines = f.readlines()
                    for v in freeOut_columns[1:]:
                        if v.split('_')[-1] == "Vol" or v.split('_')[-1] == "Surf":
                            r2l = v.split('_')[1]
                            try:
                                region_name = re.findall('[A-Z][^A-Z]*',v.split('_')[2])[0]
                            except:
                                pass
                            vol2thck = v.split('_')[3]
                            for line in lines:
                                lineList = line.split()
                                try:
                                    if lineList[4].split('-')[1][0:5] == region_name and lineList[4][0] == r2l:
                                        tempColumnDict[v] = np.float64(lineList[3])
                                    if region_name == lineList[4].split('-')[1] and lineList[4][0] == r2l:
                                        tempColumnDict[v] = np.float64(lineList[3])
                                except:
                                    pass
                                try:
                                    if v.split('_')[-1] == re.findall('[A-Z][^A-Z]*',lineList[2])[-1][:-1]:
                                        tempColumnDict[v] = np.float64(lineList[11][:-1])
                                except:
                                    pass
    tempColumnDict["Subject"] = subject
    freeOut = freeOut.append(tempColumnDict,ignore_index=True)
```
