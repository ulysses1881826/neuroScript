### This part is for MCI data

```python
## dataframe to keep the results
MCI_amyloid_pos = pd.DataFrame(columns=['PTID','EXAMDATE','ABETA_bl','AV45'])

sub_in_MM1 = list(set(list(MM1_data['Subject'])))

port = ['ADNI2','ADNI3','ADNIGO']

for ADNIindex, ADNIrow in ADNIMERGE.iterrows():
    if ADNIrow['PTID'] in sub_in_MM1 and ADNIrow['VISCODE'] == 'bl' and ADNIrow['ABETA_bl'] != '>1700' and np.float(ADNIrow['ABETA_bl']) < 1100 and ADNIrow['COLPROT'] in port:
        tempDataframe = pd.DataFrame([[ADNIrow['PTID'],ADNIrow['EXAMDATE'],ADNIrow['ABETA_bl'],ADNIrow['AV45']]],columns=['PTID','EXAMDATE','ABETA_bl','AV45'])
        MCI_amyloid_pos = MCI_amyloid_pos.append(tempDataframe)
    if ADNIrow['PTID'] in sub_in_MM1 and ADNIrow['VISCODE'] == 'bl' and ADNIrow['ABETA_bl'] != '>1700' and math.isnan(np.float(ADNIrow['ABETA_bl'])) and np.float(ADNIrow['AV45']) > 1.11 and ADNIrow['COLPROT'] in port:
        tempDataframe = pd.DataFrame([[ADNIrow['PTID'],ADNIrow['EXAMDATE'],ADNIrow['ABETA_bl'],ADNIrow['AV45']]],columns=['PTID','EXAMDATE','ABETA_bl','AV45'])
        MCI_amyloid_pos = MCI_amyloid_pos.append(tempDataframe)
        
```

```python
MCI_amyloid_pos.to_csv('MCI_amyloid_pos.csv',index=False)
```

```python
## This is for all MCI MRI data, no matter amyloid positive or amyloid negative
MCI_all = pd.DataFrame(columns=['PTID','EXAMDATE'])

sub_in_MM1 = list(set(list(MM1_data['Subject'])))

port = ['ADNI2','ADNI3','ADNIGO']

for ADNIindex, ADNIrow in ADNIMERGE.iterrows():
    if ADNIrow['PTID'] in sub_in_MM1 and ADNIrow['VISCODE'] == 'bl' and ADNIrow['COLPROT'] in port:
        tempDataframe = pd.DataFrame([[ADNIrow['PTID'],ADNIrow['EXAMDATE']]],columns=['PTID','EXAMDATE'])
        MCI_all = MCI_all.append(tempDataframe)
        
```
