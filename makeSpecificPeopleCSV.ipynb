### This part is for MCI data

```python
## dataframe to keep the results
MCI_amyloid_pos = pd.DataFrame(columns=['PTID','EXAMDATE','ABETA_bl','AV45'])

sub_in_MM1 = list(set(list(MM1_data['Subject'])))

port = ['ADNI2','ADNI3','ADNIGO']

for ADNIindex, ADNIrow in ADNIMERGE.iterrows():
    if ADNIrow['PTID'] in sub_in_MM1 and ADNIrow['VISCODE'] == 'bl' and ADNIrow['ABETA_bl'] != '>1700' and np.float(ADNIrow['ABETA_bl']) < 1100 and ADNIrow['COLPROT'] in port:
        tempDataframe = pd.DataFrame([[ADNIrow['PTID'],ADNIrow['EXAMDATE'],ADNIrow['ABETA_bl'],ADNIrow['AV45']]],columns=['PTID','EXAMDATE','ABETA_bl','AV45'])
        MCI_amyloid_pos = MCI_amyloid_pos.append(tempDataframe)
    if ADNIrow['PTID'] in sub_in_MM1 and ADNIrow['VISCODE'] == 'bl' and ADNIrow['ABETA_bl'] != '>1700' and math.isnan(np.float(ADNIrow['ABETA_bl'])) and np.float(ADNIrow['AV45']) > 1.11 and ADNIrow['COLPROT'] in port:
        tempDataframe = pd.DataFrame([[ADNIrow['PTID'],ADNIrow['EXAMDATE'],ADNIrow['ABETA_bl'],ADNIrow['AV45']]],columns=['PTID','EXAMDATE','ABETA_bl','AV45'])
        MCI_amyloid_pos = MCI_amyloid_pos.append(tempDataframe)
        
```

```python
MCI_amyloid_pos.to_csv('MCI_amyloid_pos.csv',index=False)
```

## This part is to add Amyloid and Tau PosOrNeg column

```python
ADNIMERGE_withTau = pd.read_csv(os.path.join(csvPath,'ADNIMERGE_withTau.csv'))
ADNIMERGE_withTau_AbetaTauPosNeg = ADNIMERGE_withTau
```

```python
#To be clear here, 1 represents Positive, 0 represents negative
ADNIMERGE_withTau_AbetaTauPosNeg['AbetaPosOrNeg'] = np.nan
ADNIMERGE_withTau_AbetaTauPosNeg['TauPosOrNeg'] = np.nan
ADNIMERGE_withTau_AbetaTauPosNeg['AbetaTauPosOrNeg'] = np.nan
for ADNIindex, ADNIrow in ADNIMERGE_withTau.iterrows():
    if ADNIrow['VISCODE'] == 'bl' and ADNIrow['ABETA_bl'] != '>1700' and (ADNIrow['ABETA_bl'] == '<200' or np.float(ADNIrow['ABETA_bl']) < 1100):
        ADNIMERGE_withTau_AbetaTauPosNeg['AbetaPosOrNeg'][ADNIindex] = 1
    else:
        ADNIMERGE_withTau_AbetaTauPosNeg['AbetaPosOrNeg'][ADNIindex] = 0
    if ADNIrow['VISCODE'] == 'bl' and ADNIrow['ABETA_bl'] != '>1700' and ADNIrow['ABETA_bl'] != '<200' and math.isnan(np.float(ADNIrow['ABETA_bl'])) and np.float(ADNIrow['AV45']) > 1.11:
        ADNIMERGE_withTau_AbetaTauPosNeg['AbetaPosOrNeg'][ADNIindex] = 1
    if ADNIrow['VISCODE'] == 'bl' and ADNIrow['PTAU'] != '<8' and (ADNIrow['PTAU'] == '>120' or float(ADNIrow['PTAU']) > 26.64):
        ADNIMERGE_withTau_AbetaTauPosNeg['TauPosOrNeg'][ADNIindex] = 1
    else:
        ADNIMERGE_withTau_AbetaTauPosNeg['TauPosOrNeg'][ADNIindex] = 0
    if ADNIrow['VISCODE'] == 'bl' and ADNIrow['PTAU'] != '>120' and ADNIrow['PTAU'] != '<8' and math.isnan(np.float(ADNIrow['PTAU'])) and float(ADNIrow['AV1451']) > 1.37:
        ADNIMERGE_withTau_AbetaTauPosNeg['TauPosOrNeg'][ADNIindex] = 1
    if ADNIMERGE_withTau_AbetaTauPosNeg['TauPosOrNeg'][ADNIindex] + ADNIMERGE_withTau_AbetaTauPosNeg['AbetaPosOrNeg'][ADNIindex] == 2:
        ADNIMERGE_withTau_AbetaTauPosNeg['AbetaTauPosOrNeg'][ADNIindex] = 1
    else:
        ADNIMERGE_withTau_AbetaTauPosNeg['AbetaTauPosOrNeg'][ADNIindex] = 0
    


ADNIMERGE_withTau_AbetaTauPosNeg.to_csv(os.path.join(csvPath,"ADNIMERGE_withTau_AbetaTauPosNeg.csv"))
```
